## Convolution Examples from Lectures and Tutorials

```python
from __future__ import division
import sympy as sp
from sympy.plotting import plot
from sympy import init_printing
init_printing(use_latex='mathjax')

a, b = sp.symbols('a,b', constant=True, positive=True, real=True)
t, tau = sp.symbols('t,tau', real=True)
rect = sp.Function('rect', real=True)
f = sp.Function('f', real=True)

def rect(t):
    return sp.Heaviside(t+1/2)-sp.Heaviside(t-1/2)
```

### lecture 06 example: two rect functions, equal width

```python
def f(t):
    return sp.Piecewise((0,t<=-1),
                        (sp.re(sp.integrate(rect(tau),(tau,-1/2,1/2+t))),t<=0),
                        (sp.re(sp.integrate(rect(tau),(tau,-1/2+t,1/2))),t<=1),
                        (0,True))

graph1 = plot(f(t),(t,-2,2))
```

### lecture 06 example: single-sided exponential and Heaviside

```python
def f(t):
    return sp.Piecewise((0,t<=0),
                        (sp.re(sp.integrate(sp.exp(-a*tau),(tau,0,t))),True))

graph2 = plot(f(t).replace(a,1),(t,-1,4))
```

### tutorial 2, Q2(b)

```python
def f(t):
    return sp.Piecewise((0,t<=-3/2),
                        (sp.re(sp.integrate(rect(tau/2),(tau,-1,1/2+t))),t<=-1/2),
                        (sp.re(sp.integrate(rect(tau/2),(tau,-1/2+t,1/2+t))),t<=1/2),
                        (sp.re(sp.integrate(rect(tau/2),(tau,-1/2+t,1))),t<=3/2),
                        (0,True))
        
graph4 = plot(f(t),(t,-2,2))
```
