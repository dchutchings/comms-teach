## Frequency Modulation Examples from Lectures and Tutorials

```python
from __future__ import division
import sympy as sp
from sympy.plotting import plot
import numpy as np
import scipy.fftpack as fp
import scipy.ndimage as ip
import scipy.special as sf
import matplotlib.pyplot as plt
from sympy import init_printing
init_printing(use_latex='mathjax')

t, tau, v, w = sp.symbols('t,tau,v,w', real=True)
f = sp.Function('f', real=True)

ndim = 1024
vc = 32.0
vm = 1.0
beta = 2.4
```

### Wideband FM time dependence

```python
def f(t):
    return sp.cos(vc*t+beta*sp.sin(vm*t))

plot(f(t),(t,0,3*np.pi),adaptive=False,nb_of_points=4096)
```

### Wideband FM Spectrum by
1. FFT of time dependence
2. Bessel function expressions

```python
v = np.arange(24,40)
fdat = [f(2.*np.pi*i/ndim) for i in range(ndim)]
plt.figure()
fig, ax = plt.subplots()
ax.bar(v,np.abs(fp.fft(fdat))[v])
plt.show()

spec = [sf.jv(i-vc,beta) for i in v]
plt.figure()
fig, ax = plt.subplots()
ax.bar(v,np.abs(spec)[v-v[0]])
plt.show()
```
